- #### in this graph it shows that 'AMT_INCOME_TOTAL' has an outlier but for confirming it we wil use IQR method 

```python
application_data ['AMT_INCOME_TOTAL'].describe()
```

```python
application_data.AMT_INCOME_TOTAL.quantile([0.0, 0.25, 0.5, 0.75, 0.99, 1])
```

- #### confirming using IQR method 

```python
q1= application_data.AMT_INCOME_TOTAL.quantile(0.25)
q1
```

```python
q3= application_data.AMT_INCOME_TOTAL.quantile(0.75)
q3
```

```python
# Calculate the interquartile range for the data. IQR = Q3 – Q1.
iqr= q3-q1
iqr
```

```python
# Any number less than this is a suspected outlier.
fence_low  = q1-1.5*iqr
fence_low
```

```python
# Any number greater than this is a suspected outlier.
fence_high = q3+1.5*iqr
fence_high
```

'AMT_INCOME_TOTAL' i.e income of the client has a maximum value of 117000000  and it is greater than  (q3+1.5*iqr)  so it act as an outlier which i have confirmed from IQR method.  

 - #### Reporting outlier and in 'AMT_INCOME_TOTAL' So going to deal with this variable by capping it.

```python
application_data= application_data[(application_data.AMT_INCOME_TOTAL <= 337500.0)]
application_data
```
